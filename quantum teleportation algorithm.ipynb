```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3,3) # 3 q bit, 3 c bit
```

```python
%matplotlib inline
circuit.draw(output='mpl')
```

```python
circuit.x(0) # create X atom at first q bit
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1) # create H atom at second q bit
circuit.cx(1,2) # create tanglement at second and third q bit
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0,1],[0,1])
circuit.draw(output='mpl')
# store the result to classical bit
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

```python
circuit.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
